```python
class Atm():
    
    def __init__(self):
        
        self.pin = ""
        self.balance = 0
        
        self.menu()
        
    def menu(self):
        user_input = """
        Welcome to ATM
        
        1. Create PIN
        2. Deposit Money
        3. Withdraw Money
        4. Check Balance
        5. Exit
        """
        
        if user_input == "1":
            self.create_pin()
        elif user_input == "2":
            self.deposit()
        elif user_input == "3":
            self.withdraw()
        elif user_input == "4":
            self.check_balance()
        else:
            print("Have a nice day!")
            
    def create_pin(self):
        self.pin = input("Enter your pin")
        print("Set Successfully")
    
    def deposit(self):
        temp = input("Enter your pin")
        if temp == self.pin:
            amount = int(input("Enter the Amount"))
        self.balance = self.balance + amount
        print("Deposit Successful")
    
    def withdraw(self):
        temp = input("Enter your pin")
        if temp == self.pin:
            amount  = int(input("Enter the Amount"))
            if amount < self.balance:
                self.balance = self.balance - amount
                print("Withdrawal Successful")
            else:
                print("Insufficient Funds")
        else:
            print("Invalid Pin")
        
    def check_balance(self):
        temp = input("Enter your pin")
        if temp == self.pin:
            print("Balance: ",self.balance)
        else:
            print("Invalid Pin")
```
